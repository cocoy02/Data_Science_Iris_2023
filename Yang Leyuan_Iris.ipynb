# Import data and transformation

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
data = pd.read_csv("./raw data/iris.data", sep=",", header=None)
```

```python
data.columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
```

# Modelling

```python
label = pd.DataFrame({"class": ['Iris-setosa','Iris-versicolor','Iris-virginica'],
                    "label": [0,1,2]})
```

```python
data = pd.merge(data,label, on="class",how="left")
```

## Split train and test data

```python
X = data.drop(['class','label'], axis = 1)
Y = data['label']

testsize = 0.3
seed = 0
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = testsize, random_state = seed)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

## Model Training and Performance Analysis

```python
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

### Find Optimum for KNN Classifier

```python
knn = KNeighborsClassifier()
hps = dict(n_neighbors=range(1,20))
estimator = GridSearchCV(knn, hps, return_train_score=True,  cv=5, n_jobs=5)
estimator.fit(X_train,Y_train)

print('Best hyperparameter:', estimator.best_params_)
```

### Comparison between Logistics Regression, KNN Classifier and SVC 

```python
clf = LogisticRegression()
clf.fit(X_train, Y_train)
Y_fit = clf.predict(X_test)
print("LR:", classification_report(Y_test, Y_fit))
Y_fit = clf.predict_proba(X_test)
print("LR_ROC_AUC_SCORE:",roc_auc_score(Y_test, Y_fit, multi_class="ovr"))



clf = KNeighborsClassifier(n_neighbors = estimator.best_params_['n_neighbors'])
clf.fit(X_train, Y_train)
Y_fit = clf.predict(X_test)
print("KN:", classification_report(Y_test, Y_fit))
Y_fit = clf.predict_proba(X_test)
print("KN_ROC_AUC_SCORE:",roc_auc_score(Y_test, Y_fit, multi_class="ovr"))

clf = SVC(probability=True)
clf.fit(X_train, Y_train)
Y_fit = clf.predict(X_test)
print("SVC:",  classification_report(Y_test, Y_fit))
Y_fit = clf.predict_proba(X_test)
print("SVC_ROC_AUC_SCORE:",roc_auc_score(Y_test, Y_fit, multi_class="ovr"))

```

The performance between different models are similar due to small size of data. In this case, SVC is prefered.

# Visualization Dashboard

```python
from dash import Dash, html, dcc
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output, State
from dash import dash_table as dt
from dash.exceptions import PreventUpdate
import plotly.express as px
from sklearn.neighbors import NearestNeighbors
```

## After classification, finding similar points

```python
#visit http://127.0.0.1:7777/ in your web browser.


clf = SVC()
clf.fit(X_train, Y_train)

external_stylesheets = ["https://codepen.io/chriddyp/pen/bWLwgP.css"]
app = JupyterDash(__name__, external_stylesheets=external_stylesheets)


app.layout = html.Div([
    html.H2("Similar Points within the class", style={'textAlign': 'center'}),
    html.Div([
        "Sepal Length: ",
        dcc.Input(id='sepal_length', type='number', placeholder="Input Sepal Length",min=0),
    ]),
    html.Br(),
    html.Div([
        "Sepal Width: ",
        dcc.Input(id='sepal_width', type='number', placeholder="Input Sepal Width",min=0),
    ]),
    html.Br(),
    html.Div([
        "Petal Length: ",
        dcc.Input(id='petal_length', type='number', placeholder="Input Petal Length",min=0),
    ]),
    html.Br(),
    html.Div([
        "Petal Width: ",
        dcc.Input(id='petal_width', type='number', placeholder="Input Petal Width",min=0),
    ]),
    html.Br(),
    html.Button('Submit', id='submit-val',n_clicks=0),
    html.Hr(),
    dt.DataTable(id='table',columns=[{'name': i, 'id': i}  for i in data.columns if ((i != 'index') and (i != 'level_0'))], 
                 data=[{}]),
    dcc.Graph(id="graph")
])

@app.callback(
    Output('table', 'data'),
    Input("submit-val", "n_clicks"),
    [State("sepal_width", "value"),
    State("sepal_length", "value"),
    State("petal_length", "value"),
    State("petal_width", "value")]
    
)
def number_render(n_clicks,sepal_length, sepal_width, petal_length, petal_width):
    if n_clicks is None:
        raise PreventUpdate
    if n_clicks ==0:
        raise PreventUpdate
    new_data_point = pd.DataFrame({
        'sepal_length':sepal_length, 'sepal_width':sepal_width, 'petal_length':petal_length, 'petal_width':petal_width
    }, index=[0])
    predicted_class = clf.predict(new_data_point)
   
    similar_class = data[data.label == predicted_class[0]]
    similar_class.reset_index(inplace=True, drop=True)
    X = np.array(similar_class[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
    nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree').fit(X)
    distances,indices = nbrs.kneighbors([[sepal_length, sepal_width, petal_length, petal_width]])
    dataframe = similar_class.loc[indices[0],]
    new_data_point['label'] = predicted_class[0]
    new_data_point['class'] = new_data_point.apply(lambda x: 'Iris-setosa'if x['label'] == 0 else 'Iris-versicolor' if x['label'] == 1 else 'Iris-virginica',1)
    dataframe = pd.concat([dataframe,new_data_point])
    final_result = dataframe.to_dict('records')


    return final_result

@app.callback(
    Output("graph", "figure"), 
    Input("table", "data"),
    [State("sepal_width", "value"),
    State("sepal_length", "value"),
    State("petal_length", "value"),
    State("petal_width", "value")])
def update_bar_chart(df,sepal_length, sepal_width, petal_length, petal_width):
    fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
              color='petal_length', size='petal_length', size_max=18,
               opacity=0.7)

    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=7777)

```
